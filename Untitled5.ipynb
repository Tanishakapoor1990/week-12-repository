## First feature with missing values is emp_title.
#I believe a good way to impute these values is with "Job title not given"

```python
import numpy as np 

df['emp_title'] = np.where(df['emp_title'].isnull(), 'Job title not given', df['emp_title'])
df['emp_title'].head(1000)
```

```python
filt = (df['emp_title'] ==  "Job title not given")
df.loc[filt]
```
